```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


# Load data assuming CSV files are in the same directory
distribution_centers = pd.read_csv("./data/distribution_centers.csv")
events = pd.read_csv("./data/events.csv")
inventory_items = pd.read_csv("./data/inventory_items.csv")
order_items = pd.read_csv("./data/order_items.csv")
orders = pd.read_csv("./data/orders.csv")
products = pd.read_csv("./data/products.csv")
users = pd.read_csv("./data/users.csv")
```

```python
# Age Distribution
mean_age = users['age'].mean()
median_age = users['age'].median()
mode_age = users['age'].mode()[0]

plt.figure(figsize=(8, 5))
plt.hist(users['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=1, label=f'Mean Age: {mean_age:.2f}')
plt.axvline(median_age, color='green', linestyle='dashed', linewidth=1, label=f'Median Age: {median_age}')
plt.axvline(mode_age, color='orange', linestyle='dashed', linewidth=1, label=f'Mode Age: {mode_age}')
plt.legend()
plt.show()

# 2. Gender Distribution
gender_counts = users['gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()

# 3. Location Analysis
top_states = users['state'].value_counts().head(5)
top_cities = users['city'].value_counts().head(5)
top_countries = users['country'].value_counts().head(5)

print("Top States:")
print(top_states)
print("\nTop Cities:")
print(top_cities)
print("\nTop Countries:")
print(top_countries)

# 4. Traffic Source Analysis
traffic_source_counts = users['traffic_source'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(traffic_source_counts, labels=traffic_source_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Traffic Source Distribution')
plt.axis('equal')
plt.show()
```
